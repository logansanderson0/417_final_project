###  Preprocessing of our data

```python
# import dataset 

headerList = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality']

wineData = pd.read_csv('winequality-red.csv', header=0, sep=";")

# check to see if it looks as expected
print(wineData.head())

## Note that our target variable is quality and it is an integer score between 0 and 10
```

```python
#PCA
df_pca = wineData.copy()
X_pca = df_pca.loc[:, 'fixed acidity':'alcohol']
y_pca = df_pca['quality']

X_pca.tail()

#Preprocessing in preparation for PCA: Standardizing the predictor variables
X_pca = StandardScaler().fit_transform(X_pca)

#Fit PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_pca)

X_pca.shape

plt.figure(figsize=(10,8))
plt.scatter(X_pca[:,0],X_pca[:,1],c=y_pca,cmap='rainbow', s=20)
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
plt.title("Using PCA to Visualize Classes")
plt.show()
```

### More preprocessing / exploratory data analysis

We are using statistical tests to determine which features have the best relationship with the target variable. 

Specifically, we are using the chi-squared (chi2) statistical test for non-negative features to select 10 of the best features from the Dataset.

NOTE: should we still do top 10 since theres fewer?

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X_f = wineData.loc[:, 'fixed acidity':'alcohol']
y_f = wineData['quality']

#apply SelectKBest class to extract top best features
bestFeatures = SelectKBest(score_func=chi2, k=11)
bestFeaturesFit = bestFeatures.fit(X_f,y_f)
dfscores = pd.DataFrame(bestFeaturesFit.scores_)  #Store predictor scores in a column 
dfcolumns = pd.DataFrame(X_f.columns)  #Store predictor variable names in a column

# #concatenate scores with predictor names
predScores = pd.concat([dfcolumns,dfscores],axis=1)
predScores.columns = ['Predictor','Score']   #naming the dataframe columns
print(predScores.nlargest(11,'Score'))       #print top (by score) 10 features
```

```python
## note they dropped the bottom two but we only have 1 to so maybe do more

## based on this result we shoul dprobably drop chlorides too becuase they are all so small

#Drop the bottom feature (smallest score)
selectedDf = wineData.drop('pH', 1)
selectedDf = selectedDf.drop('density', 1)

# selectedDf.dtypes
selX = selectedDf.loc[:, 'fixed acidity':'alcohol']
selY = selectedDf['quality']
```

### Modeling

```python
#Split data into training and test sets
X = wineData.loc[:, 'fixed acidity':'alcohol']
y = wineData['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state=42) #split the data
X_train.shape, y_train.shape, X_test.shape, y_test.shape


## don't think we need this
#selX_train, selX_test, selY_train, selY_test = train_test_split(selX, selY, test_size = .2, random_state=10) #split the data
#selX_train.shape, selY_train.shape, selX_test.shape, selY_test.shape
```

```python
#Use sklearn.preprocessing.StandardScaler to standardize predictors 
scaledData = StandardScaler()
X_train = scaledData.fit_transform(X_train)
X_test = scaledData.transform(X_test)

#selX_train = scaledData.fit_transform(selX_train)
#selX_test = scaledData.transform(selX_test)
```

### ANN model

```python
train_score = []
test_score = []
layers = list(range(5,300,5)) 
for i in layers:
    MLP = MLPClassifier(activation = 'logistic', 
                       solver = 'sgd', 
                       hidden_layer_sizes = (i,), 
                       alpha=1e-2, 
                       max_iter = 400,
                       learning_rate_init = 0.1)
    MLP.fit(X_train,y_train)
    train_score.append(MLP.score(X_train,y_train))
    test_score.append(MLP.score(X_test,y_test))
```

```python
# best number of layers 
layers[test_score.index(max(test_score))]
```

```python
# then use this number of layers 

MLP = MLPClassifier(activation = 'logistic', 
                       solver = 'sgd', 
                       hidden_layer_sizes = (25,), 
                       alpha=1e-2, 
                       max_iter = 400,
                       learning_rate_init = 0.1)
```

```python
MLP.fit(X_train, y_train)

y_pred = MLP.predict(X_test)

#Plot confusion matrix
cm = confusion_matrix(y_test, y_pred)
df_cm = pd.DataFrame(cm)
sns.set(font_scale=1.2) # for label size
sns.heatmap(df_cm, annot=True, annot_kws={"size": 16}) # font size
plt.show()

#Print classification report
clas = classification_report(y_test, y_pred)
print(clas)
```

### SVM model

```python
wineSVM = SVC(kernel = 'rbf')
wineSVM.fit(X_train, y_train)
y_pred = wineSVM.predict(X_test)

#Plot confusion matrix
cm = confusion_matrix(y_test, y_pred)
df_cm = pd.DataFrame(cm)
sns.set(font_scale=1.2) # for label size
sns.heatmap(df_cm, annot=True, annot_kws={"size": 16}) # font size
plt.show()

#Print classification report for rbf kernel
clas1 = classification_report(y_test, y_pred)
print("rbf kernel classification report: \n",clas1)

#Print classification report for sigmoid kernel
heartSVM = SVC(kernel = 'sigmoid')
heartSVM.fit(X_train, y_train)
y_pred = heartSVM.predict(X_test)
clas2 = classification_report(y_test, y_pred)
print("sigmoid kernel classification report: \n", clas2)
```

```python
# tuning hyperparameters 

wine_SVM = SVC()
param_grid_svm = {
    'kernel': ['rbf','sigmoid','poly'], 
    'C': [1, 10, 100, 1000]
}

CV_svm = GridSearchCV(estimator=wine_SVM, param_grid=param_grid_svm, cv= 10)
CV_svm.fit(X_train, y_train)

#print the result of best hyperparameters
print(CV_svm.best_params_)
           
```

```python
# then using the best parameters 

wineSVM_final = SVC(kernel = 'rbf', C = 10)
wineSVM_final.fit(X_train, y_train)
y_pred = wineSVM_final.predict(X_test)

print(wineSVM_final.score(X_test,y_test))

clas = classification_report(y_pred, y_test)
print(clas)
```
